### 2.2 ê¸°ë³¸ Intelligence í•¨ìˆ˜ êµ¬í˜„

````python
def basic_intelligence(prompt: str) -> str:
    """
    ì§€ëŠ¥: LLMì„ ì‚¬ìš©í•˜ì—¬ ì •ë³´ë¥¼ ì²˜ë¦¬í•˜ê³  ê²°ì •ì„ ë‚´ë¦¬ëŠ” 'ë‡Œ' ì—­í• 
    
    Args:
        prompt (str): LLMì—ê²Œ ì „ë‹¬í•  ì§ˆë¬¸ì´ë‚˜ ì§€ì‹œì‚¬í•­
        
    Returns:
        str: LLMì´ ìƒì„±í•œ ì‘ë‹µ í…ìŠ¤íŠ¸
        
    í•µì‹¬ ê°œë…:
        - ì´ê²ƒì´ AI ì—ì´ì „íŠ¸ì˜ 'ë‡Œ'ì´ë‹¤
        - ì»¨í…ìŠ¤íŠ¸ë¥¼ ì´í•´í•˜ê³  ì§€ì‹œì‚¬í•­ì„ ë”°ë¥´ë©° ì‘ë‹µì„ ìƒì„±í•œë‹¤
        - ì´ êµ¬ì„± ìš”ì†Œ ì—†ì´ëŠ” ê·¸ëƒ¥ ì¼ë°˜ ì†Œí”„íŠ¸ì›¨ì–´ì— ë¶ˆê³¼í•˜ë‹¤
    """
    try:
        # OpenAI APIë¥¼ í˜¸ì¶œí•˜ì—¬ ëª¨ë¸ë¡œë¶€í„° ì‘ë‹µì„ ìƒì„±í•œë‹¤
        response = client.chat.completions.create(
            model=MODEL,  # ì‚¬ìš©í•  ëª¨ë¸ ì§€ì •
            messages=[
                {"role": "user", "content": prompt}  # ì‚¬ìš©ì ë©”ì‹œì§€
            ]
        )
        
        # ì‘ë‹µì—ì„œ í…ìŠ¤íŠ¸ ë‚´ìš©ë§Œ ì¶”ì¶œí•˜ì—¬ ë°˜í™˜í•œë‹¤
        return response.choices[0].message.content
    except Exception as e:
        return f"API í˜¸ì¶œ ì¤‘ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤: {e}"
````

### 2.3 í•¨ìˆ˜ ì‚¬ìš© ì˜ˆì œ

### 4.2 ì‹œìŠ¤í…œ í”„ë¡¬í”„íŠ¸ í™œìš©
System í”„ë¡¬í”„íŠ¸ëŠ” AIì˜ ì—­í• , í˜ë¥´ì†Œë‚˜, í–‰ë™ ì§€ì¹¨ì„ ì •ì˜í•˜ì—¬ ì¼ê´€ì„± ìˆëŠ” ì‘ë‹µì„ ìœ ë„í•œë‹¤.

````python
def intelligence_with_system(user_prompt: str, system_prompt: str) -> str:
    """
    ì‹œìŠ¤í…œ í”„ë¡¬í”„íŠ¸ë¥¼ í™œìš©í•œ ê³ ê¸‰ Intelligence í•¨ìˆ˜
    
    Args:
        user_prompt: ì‚¬ìš©ìì˜ ì§ˆë¬¸/ìš”ì²­
        system_prompt: AIì˜ ì—­í• ê³¼ í–‰ë™ì„ ì •ì˜í•˜ëŠ” ì§€ì‹œì‚¬í•­
    """
    response = client.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ]
    )
    return response.choices[0].message.content

# ì‚¬ìš© ì˜ˆì‹œ: ì „ë¬¸ íŠœí„° ì—­í• 
system_msg = """ë‹¹ì‹ ì€ ì¹œì ˆí•˜ê³  ê²½í—˜ì´ í’ë¶€í•œ í”„ë¡œê·¸ë˜ë° íŠœí„°ì…ë‹ˆë‹¤.
ì´ˆë³´ìë„ ì´í•´í•  ìˆ˜ ìˆë„ë¡ ì‰½ê²Œ ì„¤ëª…í•˜ë©°, í•­ìƒ ì˜ˆì‹œë¥¼ í¬í•¨í•©ë‹ˆë‹¤.
ì„¤ëª…ì€ ê°„ê²°í•˜ë©´ì„œë„ í•µì‹¬ì„ ë‹´ì•„ì•¼ í•©ë‹ˆë‹¤."""

user_msg = "íŒŒì´ì¬ì˜ ë¦¬ìŠ¤íŠ¸ ì»´í”„ë¦¬í—¨ì…˜ì„ ì„¤ëª…í•´ì£¼ì„¸ìš”."

answer = intelligence_with_system(user_msg, system_msg)
print(answer)
````

### 4.3 ë‹¤ì–‘í•œ Temperature ê°’ ë¹„êµ

````python
def compare_temperatures(prompt: str):
    """ë‹¤ì–‘í•œ temperature ê°’ìœ¼ë¡œ ì‘ë‹µì„ ë¹„êµí•œë‹¤"""
    # 0.0 ëŒ€ì‹  0.1ì„ ì‚¬ìš© (ì¼ë¶€ ëª¨ë¸ì€ 0.0ì„ ì§€ì›í•˜ì§€ ì•ŠìŒ)
    temperatures = [0.1, 0.7, 1.5]
    
    for temp in temperatures:
        response = client.chat.completions.create(
            model=MODEL,
            messages=[{"role": "user", "content": prompt}],
            temperature=temp
        )
        
        print(f"\n=== Temperature: {temp} ===")
        print(response.choices[0].message.content)
        print("-" * 80)

# ì‚¬ìš© ì˜ˆì‹œ
creative_prompt = "ë¯¸ë˜ì˜ AI ê¸°ìˆ ì— ëŒ€í•œ ì§§ì€ ì´ì•¼ê¸°ë¥¼ ì¨ì£¼ì„¸ìš”."
compare_temperatures(creative_prompt)
````

### 4.4 ëŒ€í™” ì»¨í…ìŠ¤íŠ¸ ìœ ì§€
LLMì€ ê¸°ë³¸ì ìœ¼ë¡œ ìƒíƒœë¥¼ ì €ì¥í•˜ì§€ ì•ŠëŠ”ë‹¤(Stateless). ë”°ë¼ì„œ ì´ì „ ëŒ€í™” ë‚´ìš©ì„ `messages` ë¦¬ìŠ¤íŠ¸ì— ê³„ì† ì¶”ê°€í•˜ì—¬ ì „ë‹¬í•´ì•¼ ëŒ€í™”ì˜ ë§¥ë½ì„ ìœ ì§€í•  ìˆ˜ ìˆë‹¤.

````python
def conversational_intelligence(conversation_history: list) -> str:
    """
    ëŒ€í™” ê¸°ë¡ì„ ìœ ì§€í•˜ë©´ì„œ ì‘ë‹µì„ ìƒì„±í•˜ëŠ” í•¨ìˆ˜
    
    Args:
        conversation_history: ì´ì „ ëŒ€í™” ë‚´ìš©ì„ ë‹´ì€ ë¦¬ìŠ¤íŠ¸
    """
    response = client.chat.completions.create(
        model=MODEL,
        messages=conversation_history
    )
    return response.choices[0].message.content

# ì‚¬ìš© ì˜ˆì‹œ: ë‹¤ë‹¨ê³„ ëŒ€í™”
conversation = [
    {"role": "user", "content": "íŒŒì´ì¬ì—ì„œ ë¦¬ìŠ¤íŠ¸ë€ ë¬´ì—‡ì¸ê°€ìš”?"}
]

# ì²« ë²ˆì§¸ ì‘ë‹µ
first_response = conversational_intelligence(conversation)
print("ì²« ë²ˆì§¸ ì§ˆë¬¸ì— ëŒ€í•œ ë‹µë³€:")
print(first_response)
print("\n" + "="*80 + "\n")

# ëŒ€í™” ê¸°ë¡ì— LLMì˜ ì‘ë‹µ(assistant)ê³¼ ì‚¬ìš©ìì˜ ë‹¤ìŒ ì§ˆë¬¸(user)ì„ ì¶”ê°€
conversation.append({"role": "assistant", "content": first_response})
conversation.append({"role": "user", "content": "ê·¸ë ‡ë‹¤ë©´ ë¦¬ìŠ¤íŠ¸ì˜ ì£¼ìš” ë©”ì„œë“œëŠ” ë¬´ì—‡ì´ ìˆë‚˜ìš”?"})

# ë‘ ë²ˆì§¸ ì‘ë‹µ (ì»¨í…ìŠ¤íŠ¸ ìœ ì§€ë¨)
second_response = conversational_intelligence(conversation)
print("ë‘ ë²ˆì§¸ ì§ˆë¬¸ì— ëŒ€í•œ ë‹µë³€ (ì»¨í…ìŠ¤íŠ¸ ìœ ì§€):")
print(second_response)
````

## 5. ê³ ê¸‰ í™œìš© íŒ¨í„´

### 5.1 êµ¬ì¡°í™”ëœ ì¶œë ¥ ìš”ì²­
í”„ë¡¬í”„íŠ¸ì— ëª…í™•í•œ ì¶œë ¥ í˜•ì‹ì„ ì§€ì •í•˜ì—¬ (ì˜ˆ: JSON, Markdown), íŒŒì‹±í•˜ê¸° ì‰¬ìš´ í˜•íƒœë¡œ ê²°ê³¼ë¥¼ ì–»ì„ ìˆ˜ ìˆë‹¤.

````python
def get_structured_output(topic: str) -> str:
    """íŠ¹ì • í˜•ì‹ìœ¼ë¡œ êµ¬ì¡°í™”ëœ ì¶œë ¥ì„ ìš”ì²­"""
    prompt = f"""
    ì£¼ì œ: {topic}
    
    ë‹¤ìŒ Markdown í˜•ì‹ìœ¼ë¡œ ë‹µë³€í•´ì£¼ì„¸ìš”:
    
    ## ì •ì˜
    [ê°„ë‹¨í•œ ì •ì˜]
    
    ## ì£¼ìš” íŠ¹ì§•
    1. [íŠ¹ì§• 1]
    2. [íŠ¹ì§• 2]
    3. [íŠ¹ì§• 3]
    
    ## ì‹¤ì œ í™œìš© ì‚¬ë¡€
    - [ì‚¬ë¡€ 1]
    - [ì‚¬ë¡€ 2]
    
    ## ì¥ë‹¨ì 
    **ì¥ì :**
    - [ì¥ì  1]
    
    **ë‹¨ì :**
    - [ë‹¨ì  1]
    """
    return basic_intelligence(prompt)

# ì‚¬ìš© ì˜ˆì‹œ
result = get_structured_output("íŠ¸ëœìŠ¤í¬ë¨¸ ì•„í‚¤í…ì²˜")
print(result)
````

### 5.2 ì—ëŸ¬ í•¸ë“¤ë§
ë„¤íŠ¸ì›Œí¬ ë¬¸ì œ, API í‚¤ ì˜¤ë¥˜, ì‚¬ìš©ëŸ‰ ì´ˆê³¼ ë“± API í˜¸ì¶œ ì‹œ ë°œìƒí•  ìˆ˜ ìˆëŠ” ë‹¤ì–‘í•œ ì˜ˆì™¸ ìƒí™©ì— ëŒ€ë¹„í•˜ëŠ” ê²ƒì´ ì¤‘ìš”í•©ë‹ˆë‹¤.

````python
def safe_intelligence(prompt: str) -> str:
    """ì—ëŸ¬ ì²˜ë¦¬ê°€ í¬í•¨ëœ ì•ˆì „í•œ Intelligence í•¨ìˆ˜"""
    try:
        response = client.chat.completions.create(
            model=MODEL,
            messages=[{"role": "user", "content": prompt}],
            timeout=30  # 30ì´ˆ íƒ€ì„ì•„ì›ƒ ì„¤ì •
        )
        return response.choices[0].message.content
        
    except Exception as e:
        error_type = type(e).__name__
        
        if "RateLimitError" in error_type:
            return "API í˜¸ì¶œ í•œë„ë¥¼ ì´ˆê³¼í–ˆìŠµë‹ˆë‹¤. ì ì‹œ í›„ ë‹¤ì‹œ ì‹œë„í•˜ì„¸ìš”."
        elif "Timeout" in error_type:
            return "ìš”ì²­ ì‹œê°„ì´ ì´ˆê³¼ë˜ì—ˆìŠµë‹ˆë‹¤. ë‹¤ì‹œ ì‹œë„í•˜ì„¸ìš”."
        elif "AuthenticationError" in error_type:
            return "API í‚¤ ì¸ì¦ì— ì‹¤íŒ¨í–ˆìŠµë‹ˆë‹¤. API í‚¤ë¥¼ í™•ì¸í•˜ì„¸ìš”."
        else:
            return f"ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤: {str(e)}"

# ì‚¬ìš© ì˜ˆì‹œ
result = safe_intelligence("ì•ˆë…•í•˜ì„¸ìš”!")
print(result)
````

### 5.3 í† í° ì‚¬ìš©ëŸ‰ ì¶”ì 
API ì‘ë‹µ ê°ì²´ì—ëŠ” í† í° ì‚¬ìš©ëŸ‰ ì •ë³´(`usage`)ê°€ í¬í•¨ë˜ì–´ ìˆì–´ ë¹„ìš©ì„ ê´€ë¦¬í•˜ê³  ëª¨ë‹ˆí„°ë§í•˜ëŠ” ë° ìœ ìš©í•©ë‹ˆë‹¤.

````python
def intelligence_with_usage(prompt: str) -> dict:
    """í† í° ì‚¬ìš©ëŸ‰ì„ í•¨ê»˜ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜"""
    response = client.chat.completions.create(
        model=MODEL,
        messages=[{"role": "user", "content": prompt}]
    )
    
    return {
        "content": response.choices[0].message.content,
        "usage": {
            "prompt_tokens": response.usage.prompt_tokens,
            "completion_tokens": response.usage.completion_tokens,
            "total_tokens": response.usage.total_tokens
        }
    }

# ì‚¬ìš© ì˜ˆì‹œ
result = intelligence_with_usage("AI ì—ì´ì „íŠ¸ì˜ 7ê°€ì§€ êµ¬ì„± ìš”ì†Œë¥¼ ë‚˜ì—´í•´ì£¼ì„¸ìš”.")
print("ì‘ë‹µ:", result["content"])
print(f"\ní† í° ì‚¬ìš©ëŸ‰:")
print(f"  - í”„ë¡¬í”„íŠ¸: {result['usage']['prompt_tokens']} í† í°")
print(f"  - ì‘ë‹µ: {result['usage']['completion_tokens']} í† í°")
print(f"  - ì´í•©: {result['usage']['total_tokens']} í† í°")
````
